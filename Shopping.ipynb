```python
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

# Importing Dataset

```python
dataset = []
with open('Market_Basket_Optimisation.csv') as file:
  reader =  csv.reader(file, delimiter=',')
  for row in reader:
    dataset += [row]
```

```python
dataset[0:11]
```

```python
len(dataset)
```

# Processing Data

```python
te = TransactionEncoder()
x = te.fit_transform(dataset)
```

```python
x
```

```python
te.columns_ #Model Parameter that's why there is underscore
```

```python
len(te.columns_)
```

```python
df = pd.DataFrame(x, columns=te.columns_)
```

```python
df
```

# Finding Frequent itemsets

```python

freq_itemset = apriori(df, min_support=0.02, use_colnames=True)
```

```python
freq_itemset
```

# Finding the Association Rules

```python
rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.25)
rules
```

```python
cut = rules[['antecedents', 'consequents', 'support', 'confidence']] #Slicing important rules
cut
```
